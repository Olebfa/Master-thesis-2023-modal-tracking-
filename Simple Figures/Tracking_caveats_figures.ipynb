```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(context='paper',font_scale=1.0, style = 'ticks', color_codes='deep')
```

```python
np.random.seed(1)

t = np.arange(0, 15, 1)
true_f = np.ones_like(t)
f_koma = true_f + np.random.normal(loc=0, scale = 0.09, size = t.shape[0])

f_koma[11] += 0.5
f_koma[12] += 0.2
f_koma[-2] -= 0.25
```

```python
fig, ax = plt.subplots(figsize = (15, 8))

ax.plot(t, true_f, color = 'tab:red', label = 'Analytical solution')
ax.scatter(t, f_koma, color = 'C0', label = 'AOMA output')
ax.set_ylim(0.5,1.5)

ax.plot([0,14],[1.25, 1.25], c = 'tab:red', linestyle = 'dashed', label = 'Threshold')
ax.plot([0,14],[0.75, 0.75], c = 'tab:red', linestyle = 'dashed')


ax.legend()
```

```python
fig, ax = plt.subplots(figsize = (15, 8))

ax.plot(t, true_f, color = 'red', label = 'Analytical solution')
ax.scatter(t, f_koma, color = 'C0', label = 'AOMA output')
ax.set_ylim(0.5,1.5)

ax.plot([6,6], [0, 2], color = 'darkgreen', linestyle = 'dashed', label = 'Current time segment')
ax.plot(t[:6], f_koma[:6], color = 'g')

ax.annotate(text = ' ',xy=(t[6], true_f[6]), xytext=(t[6], f_koma[6]),
            arrowprops=dict(arrowstyle='<->', color='red'))

ax.annotate(text = ' ',xy=(t[5], f_koma[5]), xytext=(t[5], f_koma[6]),
            arrowprops=dict(arrowstyle='<->', color='C0'))

ax.legend()
```
