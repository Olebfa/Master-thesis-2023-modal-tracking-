```python

from matplotlib.patches import FancyArrowPatch
from mpl_toolkits.mplot3d.proj3d import proj_transform
from mpl_toolkits.mplot3d.axes3d import Axes3D
class Arrow3D(FancyArrowPatch):

    def __init__(self, x, y, z, dx, dy, dz, *args, **kwargs):
        super().__init__((0, 0), (0, 0), *args, **kwargs)
        self._xyz = (x, y, z)
        self._dxdydz = (dx, dy, dz)

    def draw(self, renderer):
        x1, y1, z1 = self._xyz
        dx, dy, dz = self._dxdydz
        x2, y2, z2 = (x1 + dx, y1 + dy, z1 + dz)

        xs, ys, zs = proj_transform((x1, x2), (y1, y2), (z1, z2), self.axes.M)
        self.set_positions((xs[0], ys[0]), (xs[1], ys[1]))
        super().draw(renderer)
        
    def do_3d_projection(self, renderer=None):
        x1, y1, z1 = self._xyz
        dx, dy, dz = self._dxdydz
        x2, y2, z2 = (x1 + dx, y1 + dy, z1 + dz)

        xs, ys, zs = proj_transform((x1, x2), (y1, y2), (z1, z2), self.axes.M)
        self.set_positions((xs[0], ys[0]), (xs[1], ys[1]))

        return np.min(zs) 
```

```python
def _arrow3D(ax, x, y, z, dx, dy, dz, *args, **kwargs):
    '''Add an 3d arrow to an `Axes3D` instance.'''

    arrow = Arrow3D(x, y, z, dx, dy, dz, *args, **kwargs)
    ax.add_artist(arrow)


```

```python
def plot_to_ax(ax,px_def,py_def):
    c='grey'
    a=0.6

    points_d = np.array([px,py,pz-0.02]).T.reshape(-1, 1, 3)
    segments_ver_d =np.concatenate([points_d[:-4], points_d[4:]], axis=1)


    lines=Line3DCollection(segments_ver_d[:4],linewidths=3,color=c,alpha=a)
    lines.set(joinstyle='round',capstyle='round')
    ax.add_collection(lines)
    
    lines_2=Line3DCollection(segments_ver_d[4:],linewidths=3,color=c,alpha=a)
    lines_2.set(joinstyle='round',capstyle='round')
    ax.add_collection(lines_2)
    
    ax.plot_trisurf(px[4:8],py[4:8],pz[4:8]-0.01,color=c,shade=True,
                    edgecolor=[0,0,0,0],alpha=a,antialiased=False,linewidth=0)
    ax.plot_trisurf(px[8:],py[8:],pz[8:]-0.01,color=c,shade=True,
                    edgecolor=[0,0,0,0],alpha=a,antialiased=False,linewidth=0)
  
    ax.set_xlim(-2,12)
    ax.set_ylim(-2,12)
    ax.set_zlim(0,21)

    color_tuple=(1,1,1,0)
    ax.w_xaxis.set_pane_color(color_tuple)
    ax.w_yaxis.set_pane_color(color_tuple)
    ax.w_zaxis.set_pane_color((0,0,0,0.2))
    ax.w_xaxis.line.set_color(color_tuple)
    ax.w_yaxis.line.set_color(color_tuple)
    ax.w_zaxis.line.set_color(color_tuple)

    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_zticks([])
    
    ax.set_xlabel('<--- X',labelpad=-15)
    ax.set_ylabel('Y --->',labelpad=-15)
    
    a=0.9
    px_d=px+px_def
    py_d=py+py_def

    points_d = np.array([px_d,py_d,pz]).T.reshape(-1, 1, 3)
    segments_ver_d =np.concatenate([points_d[:-4], points_d[4:]], axis=1)
    # print(segments_ver_d_1)
    
   
    c='tab:red'
    # c='grey'
    lines_d=Line3DCollection(segments_ver_d[:4],linewidths=5,zorder=1,color=c,alpha=a)
    lines_d.set(joinstyle='round',capstyle='round')
    ax.add_collection(lines_d)
    
    lines_d2=Line3DCollection(segments_ver_d[4:],linewidths=5,zorder=1,color=c,alpha=a)
    lines_d2.set(joinstyle='round',capstyle='round')
    ax.add_collection(lines_d2)
    
    ax.plot_trisurf(px_d[4:8],py_d[4:8],pz[4:8],color=c,zorder=5,
                    edgecolor=[0,0,0,0],alpha=1,antialiased=False,linewidth=0)
    ax.plot_trisurf(px_d[8:],py_d[8:],pz[8:],color=c,zorder=5,
                    edgecolor=[0,0,0,0],alpha=1,antialiased=False,linewidth=0)

    # ax.set_axis_off()
    return ax
```
