```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets as data
%matplotlib inline
sns.set_theme(context='paper', style='ticks',color_codes='deep')
# plot_kwds = {'alpha' : 0.8, 's' : 100, 'linewidths':0}
```

```python
map=sns.color_palette("blend:b,orangered", as_cmap=True)
# sns.color_palette("blend:#7AB,#EDA", as_cmap=True)
```

```python
np.random.seed(2)
mock_data, _ = data.make_blobs(n_samples=10, centers=[(1,1)], cluster_std=0.3)
p1 = mock_data[5,:] #outer small
p2 = mock_data[1,:] #center big
p3 = mock_data[2,:] #outer big
p4 = mock_data[0,:] #center small
a=1
si=10
circle2 = plt.Circle((mock_data[1,:]), np.linalg.norm(p2-p3), color='b',fill=False) #center big
circle4 = plt.Circle((mock_data[0,:]), np.linalg.norm(p1-p4), color='g',fill=False) #center small

fig, axis = plt.subplots(figsize = (3,3))
axis.scatter(mock_data.T[0], mock_data.T[1], color='b')
axis.add_patch(circle2)
axis.add_patch(circle4)
axis.axis('equal')
axis.plot([p2[0], p4[0]],[p2[1], p4[1]], color = 'orange',alpha=a)
axis.plot([p1[0], p4[0]],[p1[1], p4[1]], color ='g',alpha=a)
axis.plot([p2[0], p3[0]],[p2[1], p3[1]], color = 'b',alpha=a)
axis.annotate('$x_{p}$', xy = p4 + [0.02,0.02], color = 'r',alpha=a, size = si)
axis.annotate('$x_{q}$', xy = p2 + [-0.11,0.03], color = 'r',alpha=a, size =si)
axis.annotate('$d_{core}(x_{p})$', xy = (0.49, 0.83), size = si, color = 'g',alpha=a)
axis.annotate('$d_{core}(x_{q})$', xy = (0.5, 1.63), size = si, color = 'b',alpha=a)
axis.annotate('$d(x_{p},x_{q})$', xy = (0.6, 1.25), size = si, color = 'orange',alpha=a)
axis.set_xlim([0,1.25])
axis.set_ylim([0.5,1.80])
axis.set_xticks(np.arange(0, 1.26, 0.25))
axis.set_yticks(np.arange(0.5, 1.81, 0.2))
axis.set_xlabel('Metric 1')
axis.set_ylabel('Metric 2')
plt.savefig(fname='MRD.png',dpi=400,bbox_inches='tight')
```

```python
np.random.seed(1)
#moons, _ = data.make_moons(n_samples=50, noise=0.05)
test_data, _ = data.make_blobs(n_samples=75, centers=[(-0.75,2.25), (1.0, 2.0), (0.0, 1.0)], cluster_std=0.30)
#test_data = np.vstack([moons, blobs])


fig, axs = plt.subplots(1,2,figsize = (7,3), gridspec_kw={'wspace':0.4})

axs[0].scatter(test_data.T[0], test_data.T[1], color='b')

clusterer.minimum_spanning_tree_.plot(axis = axs[1], edge_cmap=map,#'brg',
                                      edge_alpha=1,
                                      node_size=40,
                                      edge_linewidth=1.7
        )

axs[0].set_title('Example data set')
axs[1].set_title('Mutual reachability graph')

# fig.colorbar(location='bottom')''
fig.savefig('./dataset_mutualgraph.png',dpi=400, bbox_inches = 'tight')                               
```

```python
fig, axs = plt.subplots(1,2,figsize = (7,3), gridspec_kw={'wspace':0.4})
clusterer.single_linkage_tree_.plot(axis = axs[0], cmap=map,vary_line_width=True,colorbar=True)
clusterer.condensed_tree_.plot(axis = axs[1], cmap = map)#'viridis')

axs[0].set_title('Dendrogram')
axs[1].set_title('Condensed dendrogram')

plt.savefig(fname='dendrogram_condenseddendrogram.png',dpi=400, bbox_inches = 'tight')
```

```python
fig, axs = plt.subplots(1,2,figsize = (7,3), gridspec_kw={'wspace':0.4})
clusterer.condensed_tree_.plot(axis = axs[0], select_clusters=True, selection_palette=sns.color_palette(), cmap = map)

palette = sns.color_palette()
cluster_colors = [sns.desaturate(palette[col], sat)
                  if col >= 0 else (0.5, 0.5, 0.5) for col, sat in
                  zip(clusterer.labels_, clusterer.probabilities_)]

axs[1].scatter(test_data.T[0], test_data.T[1], c=cluster_colors, )


plt.savefig(fname='condensenddendrogram_selectedclusters.png',dpi=400, bbox_inches = 'tight')
```
