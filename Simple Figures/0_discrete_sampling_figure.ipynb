```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
sns.set_theme(context='paper', style = 'ticks', color_codes='deep')
# plot_kwds = {'alpha' : 0.8, 's' : 100, 'linewidths':0}
```

```python
t = np.linspace(0,5,100)
y = np.sin(t)

td = np.linspace(0,5,16)
yd = np.sin(td)

fig, axs = plt.subplots(figsize = (7,2))
axs.plot(t, y, color = 'g', label='True signal')
axs.scatter(td, yd, color = 'r',  label='Samples')
axs.set_xlabel('$t$ (s)')
for k in range(yd.shape[0]-1):
    if k == 0:
        axs.plot([td[k], td[k+1]], [yd[k], yd[k]], color = 'r', label='Estimated signal')
        axs.plot([td[k+1], td[k+1]], [yd[k], yd[k+1]], color = 'r')
    else:
        axs.plot([td[k], td[k+1]], [yd[k], yd[k]], color = 'r')
        axs.plot([td[k+1], td[k+1]], [yd[k], yd[k+1]], color = 'r')
axs.legend()
plt.tight_layout()
plt.savefig(fname='ZOH.png',dpi=400,bbox_inches='tight')
```

```python
fig,ax=plt.subplots(figsize = (7,1.5))
seg=np.arange(0,100)
graph=1+1/np.log(seg)
ax.plot(seg,graph,color='r')
ax.set_xlabel('Time segment')
ax.set_ylabel(r'$1+\dfrac{1}{log(h)}$')

plt.savefig(fname='cabboi_threshold.png',dpi=400,bbox_inches='tight')
```
