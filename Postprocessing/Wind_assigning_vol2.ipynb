```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import sys
import os
src_path='../src'
if not src_path in sys.path:
    sys.path.insert(1,src_path)

from preposessing import *
from utils_OB import *
from _tracking import *
%load_ext autoreload
import pickle
%autoreload 2
import pandas as pd
from utils_JK import *
from _postprocessing import *
import seaborn as sns
sns.set_theme(context='paper', style = 'ticks', color_codes='deep')
matplotlib.use('module://matplotlib_inline.backend_inline')
```

```python
wheater_path='../../../A6_wind/'

mean_wheater={}
for file_name in os.listdir(wheater_path):
    wheater=import_converted_ts(wheater_path,file_name)
    h=np.nanmean(wheater[0]) 
    v=np.nanmean(wheater[1]) 
    a=np.nanmean(wheater[2]) 
    T=np.nanmean(wheater[3])
    mean_wheater[file_name[:-4]]=[h,v,a,T]


freqs_damp={}
damp_path='../../../5hz_resampled_int_trans_clustered/'
for file_name in os.listdir(damp_path):
    ts=import_converted_ts(damp_path,file_name)
    dampings=np.asarray(ts.xi_auto,dtype=object)
    dampings_median=[np.median(c) for c in ts.xi_auto]
    freqs_damp[file_name[:-4]]=ts.lambds_median,dampings_median
```

```python
key=list(freqs_damp.keys())[0]
print(freqs_damp.get(key))
```

```python
traces_path='../../../Tracking/Method1_a_n10/'

trace_names=os.listdir(traces_path) 

for name in trace_names:
    trace=import_converted_ts(traces_path,name)
    trace.vertical=[]
    trace.horisontal=[]
    trace.temp=[]
    trace.angle=[]
    trace.damping=[]
    trace.freq_comparison=[]
    count_good=0
    count_fail=0
    for i,segment in enumerate(trace.file_names):
        try:
            h=mean_wheater[segment][0]
            v=mean_wheater[segment][1]
            a=mean_wheater[segment][2]
            T=mean_wheater[segment][3]
            fs=freqs_damp[segment][0]
            xis=freqs_damp[segment][1]
            this_f=trace.frequencies[i]
            
            ix=np.argmin(np.abs(np.asarray(fs)-this_f*2*np.pi))
            print(ix)
            xi=xis[ix]
            
            trace.vertical.append(v)
            trace.horisontal.append(h)
            trace.temp.append(T)
            trace.angle.append(a)
            trace.damping.append(xi)
            trace.freq_comparison.append(trace.frequencies[i])
        
            count_good+=1
        except:
            count_fail+=1
    print('Trace')        
    print(  'Successful : ',count_good, '  Failed: ',count_fail

          )
    

    with open(traces_path+name, 'wb') as fout:
        pickle.dump(trace, fout)
```
