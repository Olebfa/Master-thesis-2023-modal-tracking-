```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import sys
import os
src_path='../src'
if not src_path in sys.path:
    sys.path.insert(1,src_path)
from preposessing import *
from utils_OB import *
from _tracking import *
%load_ext autoreload
import pickle
%autoreload 2
import pandas as pd
from utils_JK import *
from _postprocessing import *
import seaborn as sns
sns.set_theme(context='paper', style = 'ticks', color_codes='deep')
matplotlib.use('module://matplotlib_inline.backend_inline')
```

```python
trace_comparisons_alg1=PostTraceComparison()
trace_comparisons_alg1.import_Post_traces_from_folder('../../../Tracking/Method1_a_n8_whether_test/')
```

```python
names=np.asarray(os.listdir('../../../5hz_clustered_int/'))
good_names=[]
for name in names:
    good_names.append(name[:-4])
print(good_names[0])

fig,axs=plt.subplots(4,4,figsize=(7,4))
count=1
for i,row in enumerate(axs):
    for j,ax in enumerate(row): 
        ptrace=trace_comparisons_alg1.Post_traces[count-1]

        
        ix=np.where(np.isin(np.asanyarray(ptrace.file_names),good_names))
        # print(len(ix[0]))
    
        wind=np.asanyarray(ptrace.horisontal)[ix]
        damp=np.asanyarray(ptrace.dampings)[ix]
        ax.scatter(wind,damp,s=1,alpha=0.3)
        
        corrcoef=np.corrcoef(wind,damp)[1,0]
        df=pd.DataFrame([wind,damp]).T
        corrcoef=df.corr()[0][1]
        # print(df.corr())
        
        try:
            ax.set_xlim(np.nanmean(wind)-3*np.nanstd(wind),np.nanmean(wind)+3*np.nanstd(wind))
            ax.set_ylim(np.nanmean(damp)-3*np.nanstd(damp),np.nanmean(damp)+3*np.nanstd(damp))
        except:
            # ax.set_ylim(0,20)
            pass

        ax.set_title('Trace '+str(count),pad=-15)
        co=f'{corrcoef:.2}'
        x=np.nanmean(wind)-2.5*np.nanstd(wind)
        y=np.nanmean(damp)+2.5*np.nanstd(damp)
        # x=ax.get_xlim()[1]*0.05
        # y= ,ax.get_ylim()[1]*0.95
        ax.text(x,y,str(co),ha='left',va='top',fontsize=8)  
        count+=1
        ax.set_xticks([])
        ax.set_yticks([])
        if j==0:
            ax.set_ylabel('Damping')
        if i==3:
            ax.set_xlabel('Mean wind')
    fig.suptitle('Mean wind plottet against damping for the modes')
    fig.subplots_adjust(wspace=0.1,hspace=0.3)
fig.savefig(fname='damping-wind_plot.png',dpi=800)
plt.show()
```
